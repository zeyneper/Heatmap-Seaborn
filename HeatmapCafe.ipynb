```python
import seaborn as sns

from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.rc('xtick', labelsize=14)
plt.rc('ytick', labelsize=14)
```

```python
xlabels = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
ylabels = ['Americano', 'Latte', 'Cold brew', 'Macchiato','Mocha', 'Cappuccino']

orders = np.array([[44, 31, 50, 41, 30, 21, 33],
                   [21, 12, 28, 38, 35, 43, 34],
                   [15, 18, 20, 14, 12, 33, 19],
                   [55, 20, 40, 28, 11, 24, 37],
                   [38, 29, 39, 26, 15, 36, 41],
                   [41, 44, 19, 18, 45, 21, 34]]
                 )
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(orders,
            cmap='YlOrBr',
            vmin=0,
            xticklabels=xlabels,
            yticklabels=ylabels,
            annot=True,
            alpha=0,
            square=True,
            annot_kws={'fontsize':14, 'fontweight': 'bold', 'color': 'black'}
           )
plt.yticks(rotation=0)
plt.tick_params(
    which='both',
    bottom=False,
    left=False,
    labelbottom=False,
    labeltop=True)
plt.tight_layout();
```
